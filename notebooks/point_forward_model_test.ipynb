```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def compute_u(u_min, beta_xs):
    return np.sqrt(beta_xs ** 2 + u_min ** 2)
def compute_mu(us):
    return (us ** 2 + 2) / (us * np.sqrt(us ** 2 + 4))
```

```python
def test_plot(u_mins):
    figs, axes = plt.subplots(1,2, figsize = (16,8))
    for u_min in u_mins:
        # plot graph
        betas = np.arange(-2, 2, 0.001)
        us = compute_u(u_min, betas)
        mus = compute_mu(us)
        axes[1].plot(betas, mus, label="u_min = "+str(u_min))
        
        # plot circle
        eins_rad = plt.Circle((0, 0), 1, color='r', fill=False)
        axes[0].set_xlim((-2, 2))
        axes[0].set_ylim((-2, 2))
        axes[0].add_patch(eins_rad)
        axes[0].plot(betas, np.ones(betas.shape)*u_min, label="u_min = "+str(u_min))
    axes[0].legend()
    axes[1].legend()
    
    plt.show()
```

```python
test_plot([0.1, 0.2])
```
